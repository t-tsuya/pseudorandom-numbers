# Periodic Phenomena of Linear Congruential Generator

LCG (Linear Congruential Generator) is one of the methods that generates pseudorandom numbers.

Althogh this algorithm is used in the rand() function in C, it is not recommended to use because it has a period derived from modulus and generates bad pseudorandom numbers.

We will see the periodic phenomenona.

```python
import matplotlib.pyplot as plt
```

```python
def lcg (X, a, c, m):
    return (a * X + c) % m
```

If parameters a, c, m are not good, the period is clearly visible.

```python
x = []
y = []

X = 123456  # seed
for i in range(100000):
    X = lcg(X, 43, 5, 256)
    if i % 2 == 0:
        x.append(X / 256)
    else:
        y.append(X / 256)

fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(x, y)
plt.show()
```

It also works in 3D.

```python
x = []
y = []
z = []

X = 123456  # seed
for i in range(33333):
    X = lcg(X, 43, 5, 256)
    if i % 3 == 0:
        x.append(X / 256)
    elif i % 3 == 1:
        y.append(X / 256)
    else:
        z.append(X / 256)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
plt.scatter(x, y, z)
plt.show()
```

The rand() in C uses parameters a, c, m below.

```python
x = []
y = []

X = 123456  # seed
for i in range(100000):
    X = lcg(X, 1103515245, 12345, 32768)
    if i % 2 == 0:
        x.append(X / 32768)
    else:
        y.append(X / 32768)

fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(x, y, s=0.1)
plt.show()
```
